```python
#read and store data frame
surveys_df = pd.read_csv("surveys.csv")

#what is the data type?
type(surveys_df)

#what are the data types inside the data frame?
surveys_df.dtypes

#get the column ID's
surveys_df.columns

#get the first five rows of data
surveys_df.head()

#get the unique datapoints in a column
pd.unique(surveys_df['species_id'])
```

```python
#count the number of samples by species
species_counts = surveys_df.groupby('species_id')['record_id'].count()
print(species_counts)
```

```python
#Count the number of species DO
species_counts_DO = surveys_df.groupby('species_id')['record_id'].count()['DO']
print(species_counts_DO)
```
